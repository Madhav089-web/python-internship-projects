```python
import calendar
for i in range(1,10):
    for j in range(1,10):
        if(i==j or (i+j)==10):
            print("*",end="")
        else:
            print(" ",end="")
        
    print("")

```

```python
y=int(input("give any year: "))
if(calendar.isleap(y)):
    print(f"{y} is a leap year")
else:
    print(f"{y} is not a leap year")
```

```python
n= input("Give any number:")
if(n==n[::-1]):
    print(f"{n} is a palindrome")
else:
    print(f"{n} is not a palindrome")
```
