Print star using *

```python
def operation(x,y,op:str):
    match(op):
        case "+":
            print(f"{x}+{y}={x+y}")
            
        case "-":
            print(f"{x}-{y}={x-y}")
        case "%":
            print(f"{x}%{y}={x%y}")
        case "÷":
            print(f"{x}÷{y}={x/y}")
        case "×":
            print(f"{x}×{y}={x*y}")
        case "*":
            print(f"{x}*{y}={x*y}")
        case "^":
            print(f"{int(x)}^{int(y)}={int(x)^int(y)}")
        case "/":
            print(f"{int(x)}/{int(y)}={(x)/(y)}")
        case "**":
            print(f"{x}**{int(y)}={x**y}")
        case "//":
            print(f"{int(x)}//{int(y)}={(x)//(y)}")

        case _:
            print(f"Invalid Operation '{op}' !")

        

# def parse(exp:str):
#     exp= exp.replace("÷",'/')
#     exp=exp.replace("×","*")
#     return exp
    
           
           
        

           
    
```

ATM System Program

```python




accounts = {}
limits = 3





while True:
    try:
        print("\nHey User! What do you want to do ?")
        option = int(input("""(1) Withdraw Money\n
                           (2) Deposit Money\n
                           (3) Change Pin\n
                           (4) Create an  Account\n
                           (5) Exit\n"""))

        match option:

            case 1:
                username = input("Username: ")
                password = input("Password: ")

                found = None
                for uid, data in accounts.items():
                    if data["username"] == username:
                        found = (uid, data)
                        break

                if not found:
                    print("Account not found.")
                    continue

                uid, data = found



                if data["blocked"]:
                    print("Your account is blocked due to multiple wrong attempts.")


                    continue





                if password != data["password"]:
                    data["attempts"] += 1
                    if data["attempts"] >= limits:
                        data["blocked"] = True
                        print("Too many wrong attempts! Your account is now blocked!")
                    else:
                        print("Incorrect password.")


                    continue

                data["attempts"] = 0

                amount = int(input("Enter amount to withdraw: "))
                if amount <= data["balance"]:



                    data["balance"] -= amount



                    print("Withdrawn successfully! Current Balance:", data["balance"])
                else:
                    
                    print("Insufficient balance!")

            case 2:
                username = input("Username: ")


                password = input("Password: ")



                found = None
                for uid, data in accounts.items():



                    if data["username"] == username:


                        found = (uid, data)
                        break

                if not found:
                    print("Account not found!")



                    continue

                uid, data = found

                if data["blocked"]:



                    print("Your account is blocked!")
                    continue

                if password != data["password"]:
                    data["attempts"] += 1


                    if data["attempts"] >= limits:


                        data["blocked"] = True

                        print("Account blocked due to too many wrong attempts!")
                    else:
                        print("Incorrect password.")
                    continue

                data["attempts"] = 0

                amount = int(input("Enter amount to deposit: "))

                data["balance"] += amount


                print("Deposited successfully! Current Balance:", data["balance"])

            case 3:
                username = input("Username: ")

                old_pass = input("Old Password: ")

                found = None
                for uid, data in accounts.items():


                    if data["username"] == username:

                        found = (uid, data)
                        break

                if not found:


                    print("Account not found!")

                    continue

                uid, data = found

                if data["blocked"]:
                    

                    print("Your account is blocked!")
                    continue

                if old_pass != data["password"]:

                    data["attempts"] += 1




                    if data["attempts"] >= limits:



                        data["blocked"] = True


                        print("Account blocked due to too many wrong attempts!")

                    else:
                        print("Incorrect password!")
                    continue

                data["attempts"] = 0

                new_pass = input("Enter new PIN: ")


                data["password"] = new_pass



                print("PIN changed successfully!")

            case 4:
                username = input("Create Username: ")

                password = input("Create Password: ")

                user_id = str(uuid4())
                accounts[user_id] = {
                    "username": username,
                    "password": password,
                    "balance": 0,
                    "attempts": 0,
                    "blocked": False
                }

                print("Account created. Your User ID is:", user_id)

            case 5:



                print("Bye!")
                break

            case _:



                print("Invalid Option !")

    except Exception as e:
        print("An error occurred:", e)
```
